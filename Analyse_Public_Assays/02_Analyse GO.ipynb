```python
import pandas as pd
```

```python
assay_protein= pd.read_csv("assay_protein.csv")
assay_protein
```

```python
protein_gene= pd.read_csv("string_mapping.tsv", sep="\t")
protein_gene.rename(columns = {'queryItem':'protein', 'preferredName':'gene'}, inplace = True)
protein_gene = protein_gene[["protein", "gene"]]
protein_gene
```

```python
assay_protein_gene = pd.merge(assay_protein, protein_gene, how="inner")
assay_protein_gene.to_csv("assay_protein_gene.csv", index=False)
assay_protein_gene
```

```python
df = pd.read_csv("string_functional_annotations.tsv", sep="\t")
#df = df[df["category"]=="GO Process"][["#node", "term ID"]].reset_index(drop=True)
df.rename(columns = {'term ID':'GO', '#node':'gene'}, inplace = True)
df = df.drop_duplicates(keep="first")
df
```

```python
len(df[df["category"]=="GO Process"].GO.unique())
```

```python
assay_protein_gene_GO = pd.merge(assay_protein_gene, df, how="inner")
assay_protein_gene_GO.to_csv("assay_protein_gene_GO.csv", index=False)
assay_protein_gene_GO
```

```python
assay_protein_gene_GO["value"]=1
assay_protein_gene_GO ["term"]=assay_protein_gene_GO ["GO"] + ":" + assay_protein_gene_GO ["term description"]
assay_protein_gene_GO
```

```python
assay_protein_gene_GO = assay_protein_gene_GO[assay_protein_gene_GO.category.isin(['GO Process', 'GO Function', 'GO Component', 'STRING clusters',
       'Reactome', 'WikiPathways', 'TISSUES', 'COMPARTMENTS',
       'UniProt Keywords', 'Pfam', 'InterPro', 'DISEASES', 'SMART'])]
```

```python
assay_protein_gene_GO.category.unique()
```

```python
assay_GO_matrix= assay_protein_gene_GO[["assay", "term", "value"]].pivot(index=["assay"], columns=["term"]).fillna(0)
assay_GO_matrix.columns = assay_GO_matrix.columns.droplevel(0)
assay_GO_matrix = assay_GO_matrix.reset_index()
assay_GO_matrix = assay_GO_matrix.reset_index(drop="True")
assay_GO_matrix = pd.merge(assay_protein_gene, assay_GO_matrix, how="inner")
assay_GO_matrix
```
