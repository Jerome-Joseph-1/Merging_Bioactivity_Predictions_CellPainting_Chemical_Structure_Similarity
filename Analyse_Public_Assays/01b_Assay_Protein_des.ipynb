```python
import pandas as pd
```

```python
df=pd.read_csv("Assay_descriptions_88_V2.csv")
df["chembl_id"]= df["chembl_id"].apply(str)
df["PubChemAID"]= df["PubChemAID"].apply(str)
df
```

```python
assay_protein= pd.read_csv("assay_protein.csv")
assay_protein["assay"]= assay_protein["assay"].apply(str)
assay_protein["protein"]= assay_protein["protein"].apply(str)
assay_protein
```

```python
assay_protein = assay_protein[assay_protein.assay.isin(df.PubChemAID.unique())]
assay_protein.to_csv("assay_protein.csv", index=False)
assay_protein
```

```python
df=pd.read_csv("assay_full_names.csv", encoding='windows-1252')
#df = df[['chembl_id', 'PubChemAID', 'assay_organism','BioAssay Type', 'Target', 'Readout type']]
df["assay"]= df["assay"].apply(str)
df
```

```python
protein_gene= pd.read_csv("string_mapping.tsv", sep="\t")
protein_gene.rename(columns = {'queryItem':'protein', 'preferredName':'gene'}, inplace = True)
protein_gene = protein_gene[["protein", "gene"]]
protein_gene["protein"]= protein_gene["protein"].apply(str)
protein_gene["gene"]= protein_gene["gene"].apply(str)
protein_gene
```

```python
k = pd.merge(assay_protein, protein_gene )
k = pd.merge(k, df)
k.to_csv("assay_protein_gene_desc.csv", index=False)
```
