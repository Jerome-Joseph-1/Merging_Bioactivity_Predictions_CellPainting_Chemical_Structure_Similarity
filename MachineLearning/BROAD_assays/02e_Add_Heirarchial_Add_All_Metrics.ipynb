```python
#This notebook combines all metrics and heirarchial
import pandas as pd
```

```python
all_preds_test_compounds_m= pd.read_csv("Predictions_train_heldout_scaled_prob_all_assays_ensemble_merger_test_compounds.csv")
all_preds_test_compounds_h= pd.read_csv("Predictions_train_heldout_scaled_prob_all_assays_heirarchial_model_test_compounds.csv")
```

```python
all_preds_test_compounds_m
```

```python
all_preds_test_compounds_h = all_preds_test_compounds_h[["assay","InChICode_standardised","Data", 'heirarchial_pred', 'heirarchial_Correct']]
all_preds_test_compounds_h
```

```python
all_preds_test_compounds = pd.merge(all_preds_test_compounds_m,all_preds_test_compounds_h,  on=["assay","InChICode_standardised","Data"])
all_preds_test_compounds= all_preds_test_compounds.sort_values("assay").reset_index(drop=True)
all_preds_test_compounds.columns
```

```python
metrics_list_m = pd.read_csv("All_metrics.csv")
metrics_list_h = pd.read_csv("heirarchial_model_metrics.csv")
```

```python
metrics_list_m
```

```python
metrics_list_h
```

```python
metrics_list = pd.concat([metrics_list_m,metrics_list_h])
metrics_list = metrics_list.sort_values(["assay", "method"]).reset_index(drop=True)
metrics_list
```

```python
metrics_list.to_csv("All_metrics_merger_heirarchial.csv", index=False)
```
